```python
#Import library
import numpy as np
import pandas as pd
import time
```

# Function for synthetic data creation

```python
def data_creater(num_col=1,each_user_data=10,num_user=1):
    appended_data=[] #List for appended data

    for i in range(num_user):
        per_user_data=pd.DataFrame() #Per user data frame
        user_name="User_"+str(i)

        for j in range(num_col): #Randomly generatinf for each column from 0 to 1
            col_name="col_"+str(j)
            per_user_data[col_name]=list(np.random.rand(1,each_user_data)[0])

        per_user_data["normal_error"]=list(np.random.normal(0, 1, each_user_data)) #Column containing the std normal error values
        per_user_data["user_id"]=user_name

        coeff_list=list(np.random.rand(1,num_col)[0]) #Generating random coeff for each user
        per_user_data["Target"]=per_user_data["normal_error"] #Initialisation of target column to 0
        for j in range(num_col): #Creating the final target column
            per_user_data["Target"]=per_user_data["Target"]+coeff_list[j]*per_user_data["col_"+str(j)]
        per_user_data=per_user_data.drop("normal_error",axis=1)
        appended_data.append(per_user_data) #Appending the final list
        per_user_data.to_csv("C:/Users/Mainak Kundu/Desktop/Samiran/Codes/Multiprocessing/Data/Splitted_User/"+user_name+".csv")

    #Final joining the full data
    final_data = pd.concat(appended_data, axis=0)
    final_data=final_data.reset_index().drop('index', 1)
    return(final_data)
```

# File Write

```python
output_path="C:/Users/Mainak Kundu/Desktop/Samiran/Codes/Multiprocessing/Data/Single_file/full_user_data.csv"
start_time=time.time()
data_creater(num_col=3,each_user_data=100,num_user=5000).to_csv(output_path)
print("Time taken for full file write is {} sec".format(time.time()-start_time))
```
